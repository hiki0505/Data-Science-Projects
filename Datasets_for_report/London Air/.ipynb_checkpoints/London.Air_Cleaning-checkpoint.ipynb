## Impute missing values

```python
''' 
    I decided not to use Imputer library from sklearn, 
    instead I created my own class for imputing missing values 
    for different type of features
'''

from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
    
df_updated = DataFrameImputer().fit_transform(df)
```

```python
# I chose Label Encoding over One Hot encoding, because for each categorical feature (2) we have same amount of its data 
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for col in category_col:
    df_updated[col] = label_encoder.fit_transform(df_updated[col])
    
df_updated
```

## Outliers detection

```python
# Here I use statistical methods such as z_score and its respective formula
def detect_outliers(data):
    outliers = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
# For each column we detect it's outlier values, but I did not removed them.
outliers_dict = {}
for col in df_updated.columns:
    outliers_dict[col] = detect_outliers(df_updated[col].values)
```
